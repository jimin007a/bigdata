```python
# 국토교통부_표준지공시지가
# url 입력

url = 'https://api.odcloud.kr/api/15004246/v1/uddi:6c4f2ce3-5a37-4c02-969d-9651e9b90188_201712151547?page=1&perPage=10&serviceKey='
key = 'Ltj%2BrbUUnqHLfDSEdTnn0QldnDaBYbLdggDD7pE9TvR4nYod0ac8cD2ETqW%2FYAKWFDlElj%2Ba620BxtM%2Ber7hBA%3D%3D'
```

```python
import requests
#url불러오기
response = requests.get(url+key)
```

```python
contents = response.text
print(contents)
```

```python
#문자열을 json으로 변경
import json
json_ob = json.loads(contents)
print(json_ob)
print(json_ob.keys())
print(json_ob.values())
```

```python
# key를 기준으로 필요한 내용 추출
count = json_ob['currentCount']
print(count)
```

```python
# key를 기준으로 필요한 내용 추출
data = json_ob['data']
print(data)
```

```python
# key를 기준으로 필요한 내용 추출후 데이터프레임으로 변경
import pandas as pd
data_df = pd.DataFrame(json_ob['data'])
print(data_df)
data_df.to_csv('국토교통부_표준지공시지가_0_9.csv')
```
